# REGRESSÃO LINEAR POLINOMIAL
Formula: **y = b0 + b1x1 + b1x1² + b1x1³ + b1x1^n**

O objetivo da regressão polinomial é modelar a relação entre uma variável independente e uma variável dependente usando uma equação polinomial. Isso permite capturar padrões mais complexos e não lineares nos dados, indo além das limitações de modelos lineares simples. A regressão polinomial pode ser útil quando a relação entre as variáveis não é bem descrita por uma linha reta.


---


**Caso de Teste:** Descobrir se um candidato mentiu sobre o salário que recebia no seu ultimo emprego, que era de $160.000 ao ano. Dataset contem a base salarial da empresa por nivel

# PRE PROCESSAMENTO DE DADOS


```python
# Importando as Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Lendo o CSV e separando as variaveis de caracteristicas da  variavel de classe
caminho_arquivo = '/content/drive/MyDrive/Colab Notebooks/MACHINE A-Z/DATASETS/Position_Salaries.csv'
pd.set_option('display.max_columns', None) #Ver todas as colunas
dataset = pd.read_csv(caminho_arquivo, encoding='utf-8')
x = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

#Treinando o Modelo de Regressão Linear Com o Dataset Completo

```python
#Chamando a biblioteca de Regressão Linear e fazendo o treinamento
from sklearn.linear_model import LinearRegression
ln_reg = LinearRegression()
ln_reg.fit(x, y)
```

# Treinando o Modelo de Regressão Polinomial Com o Dataset Completo

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
x_poly = poly_reg.fit_transform(x)
ln_reg2 = LinearRegression()
ln_reg2.fit(x_poly, y)
```

# Viaualizando o Resultado da Regressão Linear

```python
plt.scatter(x, y, color = 'red')
plt.plot(x,ln_reg.predict(x), color = 'blue')
plt.title('Verdade ou Blefe - Reg Linear')
plt.xlabel('Nivel de Posição')
plt.ylabel('Salario')
plt.show()
```

# Visualizando o Resultado da Regressão Polinomial

```python
plt.scatter(x, y, color = 'red')
plt.plot(x,ln_reg2.predict(poly_reg.fit_transform(x)), color = 'blue')
plt.title('Verdade ou Blefe - Reg Polinomial')
plt.xlabel('Nivel de Posição')
plt.ylabel('Salario')
plt.show()
```

# Predizendo Novos Resultados

```python
print('Salario Do Funcionario Nivel 6.5 (Regressão Linear Simples):',ln_reg.predict([[6.5]]))
print('Salario Do Funcionario Nivel 6.5 (Regressão Polinomial):',ln_reg2.predict(poly_reg.fit_transform([[6.5]])))
```
