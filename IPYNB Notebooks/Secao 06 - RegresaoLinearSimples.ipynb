# REGRESSÃO LINEAR SIMPLES
Formula: **y = b0 + b1x1**

O objetivo da regressão linear é encontrar os valores de b0 e b1 que minimizam a soma dos quadrados das diferenças entre os valores observados de y e os valores previstos pela equação linear. Isso é feito por meio de técnicas estatísticas. Presupostos da Regressão Linear:

* Linearidade: Ter uma relação linear entre X e Y
* Homocedasticidade: Variação igual
* Normalidade Multivariada: Normalidade da distribuição de erro
* Independencia: Não podem haver padrão nos dados, sem uma autocorrelação, ou seja, as linhas não podem se afetar, como em uma bolsa de valores.
* Multicolinearidade: Variaveis independentes não podem se correlacionar

**Proposta do Código:** Prever o salário de um funcionário com 12 anos de experiência

# PRE PROCESSAMENTO DE DADOS


*   Importação das Bibliotecas
*   Leitura do CSV
*   Separação dos Dados



```python
# Importando as Bibliotecas
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Lendo o CSV e separando as variaveis de caracteristicas da  variavel de classe
caminho_arquivo = '/content/drive/MyDrive/Colab Notebooks/MACHINE A-Z/DATASETS/Salary_Data.csv'
pd.set_option('display.max_columns', None) #Ver todas as colunas

dataset = pd.read_csv(caminho_arquivo, encoding='utf-8')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Separando os Dados de Treino e Teste
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0)
```

#Treinando o Modelo de Regressão Linear

```python
#Chamando a biblioteca de Regressão Linear e fazendo o treinamento
from sklearn.linear_model import LinearRegression
regressao = LinearRegression()
regressao.fit(x_train, y_train)
```

# Predição Do Resultado Do Dataset de Teste

```python
y_pred = regressao.predict(x_test)
```

#Visualizando Os Resultados de Treino

```python
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train,regressao.predict(x_train), color = 'blue')
plt.title('Salario x Experiencia (Conjunto de Treino')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salario')
plt.show()
```

# Visualizando Os Resultados de Teste

```python
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train,regressao.predict(x_train), color = 'blue')
plt.title('Salario x Experiencia (Conjunto de Teste')
plt.xlabel('Anos de Experiência')
plt.ylabel('Salario')
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Avaliar o desempenho do modelo
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'MSE: {mse}')
print(f'MAE: {mae}')
print(f'R²: {r2}')
```

# Fazendo Uma Única Predição

```python
print(regressao.predict([[12]]))
```

O modelo prevê que o salário de um funcionário com 12 anos de experiência seja de 138.967,50 Dolares

**Nota importante:** Observe que o valor do recurso (12 anos) foi inserido em um par duplo de colchetes. Isso porque o método "predict" sempre espera uma matriz 2D como o formato de suas entradas. E colocar 12 em um par duplo de colchetes torna a entrada exatamente uma matriz 2D.
